```python
from bs4 import BeautifulSoup
import requests
```

```python
web_url = "https://gov.optimism.io/"
html = requests.get(web_url)
```

```python
soup = BeautifulSoup(html.text , "lxml")
# print(soup.prettify())

category_box = soup.find_all("td" , class_ = "category")
len(category_box)
```

```python
names = []
links = []
desc = []
# for i in range(len(category_box)):
#     name.append(category_box[i].find_all("h3")[i].text)

for descption_box in category_box:
    names.append(descption_box.find("h3").text.strip())
    desc.append(descption_box.find("div" , {"itemprop":"description"}).text.strip())
    links.append(f'{web_url}{descption_box.find("a")["href"]}')
```

```python
# for link in links:
#     print(link)
    
url = links[0]
html = requests.get(url)
soup = BeautifulSoup(html.text , "lxml")
print(soup.prettify())
```

```python
category_box = soup.find_all("tr" , class_ = "topic-list-item")
len(category_box)
```

```python
for link in links:
    
    html = requests.get(link)
    soup = BeautifulSoup(html.text, "lxml")
    category_box = soup.find_all("tr" , class_ = "topic-list-item")
    print(f"for {link} the articals are : {len(category_box)}")
```

```python
artical_title = []
artical_desc = []
artical_link = []
artical_views = []
artical_date = []
count  = 0
for link in links:
    
    html = requests.get(link)
    soup = BeautifulSoup(html.text, "lxml")
    category_box = soup.find_all("tr" , class_ = "topic-list-item")
    
    for i in range(len(category_box)):
        title = category_box[i].find("a" , class_ = "title")
        artical_title.append(title.text if title else "NA")
        
        desc_ = category_box[i].find("p" , class_ = "excerpt")
        artical_desc.append(desc_.text.strip() if desc_ else "NA" )
        
        link = category_box[i].find("a" , class_ = "title")
        artical_link.append(link["href"] if link else "NA")
        
        views = category_box[i].find("span" , class_ = "views")
        artical_views.append(views.text.strip() if views else "NA")
        
        date = category_box[i].find_all("td")[4]
        artical_date.append(date.text.strip() if date else "NA")
    
    count += 1
    print(f"page Scrapped : {count}")
     
```

```python
import pandas as pd

data = pd.DataFrame({"Artical Title" : artical_title , "Artical desc" : artical_desc , "Artical Link" : artical_link , "Artical Views" : artical_views , "Artical Date" : artical_date})
```
